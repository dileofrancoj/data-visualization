# Curvas de nivel

Es un concepto que resulta de la interseccin de una superficie Z = f(x,y) con un plano Z = k con k constante.

Es decir, f(x,y) = k

Cuando me tope en anlisis II con este concepto me preguntaba: Para qu? Por qu es til comprender esto? Dnde se aplica?

La respuesta no lleg de forma instantanea pero hoy en da le encuentro varios campos de aplicacin:

1. Topografa y cartografa



2. Fluidos: Se utilizan para representar lineas de igual presin (isobaras) o de igual velocidad (perfiles) en el flujo de un fluido. Es muy til para diseo de tuberias, perfiles de alas de avin, etc.

3. Lineas equipotenciales en campos elctricos y magnticos



Pero tambin es til comprenderlas en el contexto de machine learning por ejemplo:
En la visualizacin de sueprficies de costo en problemas de optimizacin o bien para analizar el comportamiento de modelos en espacios de features bidimensionales. En clasificacin binaria las curvas de nivel pueden representar la frontera de decisin del modelo mostrando como se dividen las clases en el espacio de caractersticas.


Tomemos el caso del paraboloide de ecuacin $z = x^2 + y^2$


Imaginemos que vamos a intersectarlo con el plano de ecuacin z = 2



O bien, con ms planos



Y las intersecciones entre el paraboloide y los planos resulta en esto:



En definitiva, una curva de nivel se puede expresar como una $f: R^2 \rightarrow R$. Es decir, funciones dfinidas en el plano y a cada punto del mismo, se le asigna un valor real.

Tomemos una funcin divertida:

f(x,y) = cos(-xy) + sinh(x + y + 1) - log(3+x^2)

Y la vamos a restringir a un dominio especfico:

D = [-2; 2] x [-2;2]

```python
import numpy as np
import matplotlib.pyplot as plt
xx = np.linspace(-2, 2)
yy = xx.copy()
X, Y = np.meshgrid(xx, yy)
```

```python
from numpy import cos, sinh, log
Z = cos(-10 * X * Y) + sinh(X + Y) - log((3 + X ** 2) / (1 + Y ** 2))
print(Z)
```

```python
plt.contour(X, Y, Z)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Representacin de f(x,y)')
plt.show()
```

```python
plt.contourf(X, Y, Z, 15)
plt.show()
```

```python
Z2 = X**2 + Y**2
```

```python
plt.contour(X, Y, Z2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Representacin de f(x,y)')
plt.show()
```
