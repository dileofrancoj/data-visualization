# Curvas de nivel

Es un concepto que resulta de la intersección de una superficie Z = f(x,y) con un plano Z = k con k constante.

Es decir, f(x,y) = k

Cuando me tope en análisis II con este concepto me preguntaba: ¿Para qué? Por qué es útil comprender esto? ¿Dónde se aplica?

La respuesta no llegó de forma instantanea pero hoy en día le encuentro varios campos de aplicación:

1. Topografía y cartografía



2. Fluidos: Se utilizan para representar lineas de igual presión (isobaras) o de igual velocidad (perfiles) en el flujo de un fluido. Es muy útil para diseño de tuberias, perfiles de alas de avión, etc.

3. Lineas equipotenciales en campos eléctricos y magnéticos



Pero también es útil comprenderlas en el contexto de machine learning por ejemplo:
En la visualización de sueprficies de costo en problemas de optimización o bien para analizar el comportamiento de modelos en espacios de features bidimensionales. En clasificación binaria las curvas de nivel pueden representar la frontera de decisión del modelo mostrando como se dividen las clases en el espacio de características.


Tomemos el caso del paraboloide de ecuación $z = x^2 + y^2$


Imaginemos que vamos a intersectarlo con el plano de ecuación z = 2



O bien, con más planos



Y las intersecciones entre el paraboloide y los planos resulta en esto:



En definitiva, una curva de nivel se puede expresar como una $f: R^2 \rightarrow R$. Es decir, funciones dfinidas en el plano y a cada punto del mismo, se le asigna un valor real.

Tomemos una función divertida:

f(x,y) = cos(-xy) + sinh(x + y + 1) - log(3+x^2)

Y la vamos a restringir a un dominio específico:

D = [-2; 2] x [-2;2]

```python
import numpy as np
import matplotlib.pyplot as plt
xx = np.linspace(-2, 2)
yy = xx.copy()
X, Y = np.meshgrid(xx, yy)
```

```python
from numpy import cos, sinh, log
Z = cos(-10 * X * Y) + sinh(X + Y) - log((3 + X ** 2) / (1 + Y ** 2))
print(Z)
```

```python
plt.contour(X, Y, Z)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Representación de f(x,y)')
plt.show()
```

```python
plt.contourf(X, Y, Z, 15)
plt.show()
```

```python
Z2 = X**2 + Y**2
```

```python
plt.contour(X, Y, Z2)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Representación de f(x,y)')
plt.show()
```
