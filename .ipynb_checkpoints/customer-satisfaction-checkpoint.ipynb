## ¿Qué porcentaje de las personas que asisten son mujeres y que % son hombres?

````python
sns.set_theme(style='darkgrid')
sns.countplot(data=df, x='Gender')
plt.title('Countplot de Sexo')
plt.xlabel('Genero')
plt.ylabel('Frecuencia')
plt.show()
````

````python
male = df[df['Gender'] == 'Male']
female = df[df['Gender'] == 'Female']
len_male = len(male)
len_female = len(female)
print('Porcentaje de hombres: %', len_male / len(df)* 100)
print('Porcentaje de mujeres: %', len_female / len(df)* 100)
````

### Distribución de la variable edad

````python
sns.histplot(data=df, x='Age', kde=True)
plt.title('Distribución de edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
````

2a. ¿El gasto depende de la edad del consumidor?

````python
avg_ticket = df['AverageSpend'].mean()
print("Gasto promedio $", avg_ticket)
````

````python

# Crear dos grupos basados en el gasto promedio
lower_t_avg_ticket = df[df['AverageSpend'] < avg_ticket]
greater_t_avg_ticket = df[df['AverageSpend'] >= avg_ticket]

# Agregar una columna para indicar el grupo
lower_t_avg_ticket['Group'] = 'Below Average Spend'
greater_t_avg_ticket['Group'] = 'Above Average Spend'

# Concatenar los dos grupos en un solo DataFrame
combined_df = pd.concat([lower_t_avg_ticket, greater_t_avg_ticket])
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Age', data=combined_df)
plt.title('Distribución de edades agrupadas por gasto')
plt.xlabel('Grupo')
plt.ylabel('Age')
plt.show()
````

¿Como se distribuye la variable Income?

````python
# sns.histplot(data=df, x='Income')
sns.kdeplot(data=df, x='Income', bw_adjust=1, color='red')
plt.title('Distirbución de la variable Income')
plt.xlabel('Ingreso')
plt.ylabel('Frecuencia')
plt.show()
````

````python
# Calcular correlación entre Income y Age
correlation = df['Income'].corr(df['Age'])
print(f"Correlación entre la varible ingreso y edad para este grupo: {correlation}")
````

### Como varía el AverageSpend en función del GroupSize

````python
df_grouped_by_group_size = df.groupby('GroupSize')
avg_spend_by_group = df_grouped_by_group_size['AverageSpend'].mean().reset_index()

print("Promedio gastado por grupo", avg_spend_by_group)
````

### Observación de este comportamiento

Estos valores no me terminan de convencer ya que es raro pensar que un grupo de 9 personas en promedio gaste lo mismo que un grupo de 2 o 3 personas. Quiero analizar la varianza que tiene el conjunto ordenado por grupos. Se esperaría que tengan varianzas distintas o que el cardinal de las muestras estratificadas sea menor.

````python
df_grouped_by_group_size['AverageSpend'].describe()
````

Y evidentemente no. Llegado a este punto dudo del dataset :P

````python
plt.figure(figsize=(12, 6))
plt.bar(avg_spend_by_group['GroupSize'], avg_spend_by_group['AverageSpend'])
plt.xlabel('Tamaño del grupo')
plt.ylabel('Promedio gastado')
plt.title('Promedio gastado por Grupo')
plt.xticks(avg_spend_by_group['GroupSize'])
plt.show()
````

### 6. Cuantos pedidos son por delivery y cuantos en el local? Como es el ingreso en ambos casos?


````python
# Income por delivery
income_delivery = df[df['DeliveryOrder'] == 1]['AverageSpend'].sum()
income_no_delivery = df[df['DeliveryOrder'] == 0]['AverageSpend'].sum()
print('Income no delivery: $', income_no_delivery)
print('Income delivery: $', income_delivery)

income_data = {
    'DeliveryType': ['No Delivery', 'Delivery'],
    'Income': [income_no_delivery, income_delivery]
}
income_df = pd.DataFrame(income_data)
````

````python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

ax = sns.countplot(data=df, x='DeliveryOrder', ax=axes[0])
axes[0].set_title('Nro. Pedidos')
axes[0].set_xlabel('Delivery')
axes[0].set_ylabel('Frecuencia')
axes[0].set_xticklabels(['No', 'Si'])

sns.barplot(data=income_df, x='DeliveryType', y='Income', ax=axes[1])
axes[1].set_title('Monto por pedidos (Delivery / No Delivery)')
axes[1].set_xlabel('Delivery')
axes[1].set_ylabel('Monto en $')
axes[1].set_xticklabels(['No', 'Si'])

plt.tight_layout()
plt.show()
````

Respecto al delivery se podría ver cual es el tipo de comida que más se pide tanto en el local como mediante delivery

````python
plt.figure(figsize=(12, 6))

sns.countplot(data=df, x='DeliveryOrder', hue='PreferredCuisine')
plt.title('Nro. Pedidos')
plt.xlabel('Delivery')
plt.ylabel('Frecuencia')
plt.xticks([0, 1], ['No', 'Sí'])

plt.tight_layout()
plt.show()
````
